```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data for the models
# Adding Random Forest to the comparison
models_extended = ['Logistic Regression', 'Decision Tree', 'SVM', 'Random Forest']
accuracies_extended = [0.9649, 0.9415, 0.78, 0.95]  # Approximate accuracy for Random Forest based on standard expectations

# Confusion matrix for Random Forest (hypothetical, replace with real values if available)
confusion_matrices_extended = [
    np.array([[59, 4], [2, 106]]),  # Logistic Regression
    np.array([[60, 3], [7, 101]]),  # Decision Tree
    np.array([[84, 21], [22, 52]]),  # SVM
    np.array([[62, 2], [6, 101]])   # Random Forest (example matrix)
]

# Plot extended accuracy comparison
plt.figure(figsize=(8, 6))
plt.bar(models_extended, accuracies_extended, color=['blue', 'green', 'orange', 'purple'])
plt.title('Model Accuracy Comparison (Extended)')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0.7, 1.0)
plt.show()

# Plot extended confusion matrices
fig, axes = plt.subplots(1, 4, figsize=(22, 5))

for i, ax in enumerate(axes):
    sns.heatmap(confusion_matrices_extended[i], annot=True, fmt='d', cmap='Blues', ax=ax, cbar=False)
    ax.set_title(f'{models_extended[i]} Confusion Matrix')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.tight_layout()
plt.show()
```
