```python
import pandas as pd

ds = pd.read_csv("gym_members_exercise_tracking.csv")
ds.info()
ds.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

ds['Gender_Encoded'] = label_encoder.fit_transform(ds['Gender'])
ds['Workout_Type_Encoded'] = label_encoder.fit_transform(ds['Workout_Type'])

data_cleaned = ds.drop(['Gender', 'Workout_Type'], axis=1)

data_cleaned.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler


scaler = StandardScaler()

data_scaled = scaler.fit_transform(ds[['Experience_Level']])
    
df_scaled = pd.DataFrame(data_scaled, columns=['Experience_Level'])

df_scaled
```
